```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
from scipy.stats import ttest_ind

# Load the data
df = pd.read_csv("Survey on AI and Personalized Advertising Ethics (Responses) - Form responses 1.csv")

# --- Data Cleaning and Preprocessing ---

# Rename columns for easier use
df.columns = [
    "Timestamp", "Age", "Occupation", "Awareness", "InteractionFrequency",
    "Platforms", "PastSpending", "ConcernLevel", "DataUsed", "Consent",
    "TransparencyExpectation", "ManipulationAgreement", "Persuasiveness",
    "FeltManipulated", "BiasNoticed", "BiasDetectionUsefulness",
    "PrivacyToolUsage", "CrossPlatformTracking", "EmotionalInfluence",
    "ControlOverAds", "ReinforceStereotypes"
]

# Convert Timestamp to datetime objects (if needed for time-series analysis, which isn't central to the hypotheses)
df['Timestamp'] = pd.to_datetime(df['Timestamp'], format="%d/%m/%Y %H:%M:%S")

# --- Exploratory Data Analysis (EDA) and Visualizations ---

# 1. Age Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Age", order=df['Age'].value_counts().index)
plt.title("Distribution of Respondents by Age Group")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Occupation Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y="Occupation", order=df['Occupation'].value_counts().index)
plt.title("Distribution of Respondents by Occupation")
plt.tight_layout()
plt.show()

# 3. Awareness of AI in Advertising
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Awareness", order=df['Awareness'].value_counts().index)
plt.title("Awareness of AI Use in Personalized Advertising")
plt.tight_layout()
plt.show()

# 4. Concern Level Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="ConcernLevel", order=sorted(df['ConcernLevel'].unique()))
plt.title("Distribution of Concern Levels about Data Collection")
plt.xlabel("Concern Level (1-5, 1=Least Concerned, 5=Most Concerned)")
plt.tight_layout()
plt.show()

# 5. Transparency Expectation Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y="TransparencyExpectation", order=df['TransparencyExpectation'].value_counts().index)
plt.title("Distribution of Transparency Expectations")
plt.tight_layout()
plt.show()

# 6.  Emotional Influence
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='EmotionalInfluence', order=sorted(df['EmotionalInfluence'].unique()))
plt.title('Distribution of Emotional Influence of Ads')
plt.xlabel("Emotional Influence Level (1-5)")
plt.tight_layout()
plt.show()

# 7. Control Over Ads
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='ControlOverAds', order=df['ControlOverAds'].value_counts().index)
plt.title('Distribution of Perceived Control Over Ads')
plt.tight_layout()
plt.show()

# 8. Reinforce Stereotypes
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='ReinforceStereotypes', order=df['ReinforceStereotypes'].value_counts().index)
plt.title('Distribution of Belief that Ads Reinforce Stereotypes')
plt.tight_layout()
plt.show()


# --- Hypothesis Testing ---

# --- Hypothesis 1: The use of AI in personalized advertising leads to significant ethical concerns regarding privacy, fairness, and transparency compared to traditional advertising methods. ---

# We'll test parts of this hypothesis using chi-squared tests and t-tests.

# 1.a:  Privacy Concerns (Association between Awareness and Concern Level)
#       Null Hypothesis (H0): There is no association between awareness of AI use and concern level.
#       Alternative Hypothesis (H1): There is an association.

contingency_table_awareness_concern = pd.crosstab(df['Awareness'], df['ConcernLevel'])
chi2_awareness, p_awareness, _, _ = chi2_contingency(contingency_table_awareness_concern)
print(f"Chi2 Statistic (Awareness vs. Concern): {chi2_awareness:.2f}")
print(f"P-value (Awareness vs. Concern): {p_awareness:.3f}")
if p_awareness < 0.05:
    print("Reject the null hypothesis: There is a statistically significant association between awareness and concern level.")
else:
    print("Fail to reject the null hypothesis.")


# 1.b: Fairness (Association between BiasNoticed and BiasDetectionUsefulness)
contingency_table_bias = pd.crosstab(df['BiasNoticed'], df['BiasDetectionUsefulness'])
chi2_bias, p_bias, _, _ = chi2_contingency(contingency_table_bias)
print(f"\nChi2 Statistic (Bias Noticed vs. Bias Detection): {chi2_bias:.2f}")
print(f"P-value (Bias Noticed vs. Bias Detection): {p_bias:.3f}")
if p_bias < 0.05:
    print("Reject the null hypothesis: There is a statistically significant association between noticing bias and believing in bias detection mechanisms.")
else:
    print("Fail to reject the null hypothesis.")


# 1.c: Transparency (Compare TransparencyExpectation levels between those aware and unaware of AI use)
# Convert TransparencyExpectation to numerical values for t-test (where possible and meaningful)
transparency_mapping = {
    "No need for detailed transparency as long as ads are relevant": 1,
    "Some transparency, but only for sensitive data": 3,
    "Full transparency about how data is collected and used": 5
}
df['TransparencyExpectation_Numeric'] = df['TransparencyExpectation'].map(transparency_mapping)
df_aware = df[df['Awareness'] == 'Yes']
df_unaware = df[df['Awareness'].isin(['No', 'Not Sure'])]  # Combine 'No' and 'Not Sure'

# Perform t-test only if we have numeric values for both groups
if not df_aware['TransparencyExpectation_Numeric'].isnull().all() and not df_unaware['TransparencyExpectation_Numeric'].isnull().all():
    t_stat_transparency, p_transparency = ttest_ind(df_aware['TransparencyExpectation_Numeric'].dropna(),
                                                    df_unaware['TransparencyExpectation_Numeric'].dropna(),
                                                    equal_var=False)  # Welch's t-test
    print(f"\nT-statistic (Transparency Expectation - Aware vs. Unaware): {t_stat_transparency:.2f}")
    print(f"P-value (Transparency Expectation - Aware vs. Unaware): {p_transparency:.3f}")
    if p_transparency < 0.05:
        print("Reject the null hypothesis: There is a statistically significant difference in transparency expectations between those aware and unaware of AI use.")
    else:
        print("Fail to reject the null hypothesis.")
else:
    print("\nCannot perform t-test for Transparency Expectation due to non-numeric or missing values.")


# --- Hypothesis 2: Incorporating bias detection and mitigation tools in AI algorithms will reduce the potential for manipulative advertising practices. ---

# 2.a Association between BiasDetectionUsefulness and ManipulationAgreement
contingency_table_manipulation = pd.crosstab(df['BiasDetectionUsefulness'], df['ManipulationAgreement'])
chi2_manipulation, p_manipulation, _, _ = chi2_contingency(contingency_table_manipulation)
print(f"\nChi2 Statistic (Bias Detection vs. Manipulation Agreement): {chi2_manipulation:.2f}")
print(f"P-value (Bias Detection vs. Manipulation Agreement): {p_manipulation:.3f}")
if p_manipulation < 0.05:
    print("Reject the null hypothesis: There is a statistically significant association between believing in bias detection and agreeing that personalized ads can be manipulative.")
else:
    print("Fail to reject the null hypothesis.")

# 2.b Association between BiasDetectionUsefulness and FeltManipulated
contingency_table_felt_manipulated = pd.crosstab(df['BiasDetectionUsefulness'], df['FeltManipulated'])
chi2_felt_manipulated, p_felt_manipulated, _, _ = chi2_contingency(contingency_table_felt_manipulated)
print(f"\nChi2 Statistic (Bias Detection vs. Felt Manipulated): {chi2_felt_manipulated:.2f}")
print(f"P-value (Bias Detection vs. Felt Manipulated): {p_felt_manipulated:.3f}")
if p_felt_manipulated < 0.05:
    print("Reject the null hypothesis: There is a statistically significant association between believing in bias detection and having felt manipulated by ads.")
else:
    print("Fail to reject the null hypothesis.")


# --- Hypothesis 3: Increasing transparency and simplifying consent mechanisms will lead to higher user trust and acceptance of personalized advertising. ---

# 3.a: Association between TransparencyExpectation and Consent
contingency_table_transparency_consent = pd.crosstab(df['TransparencyExpectation'], df['Consent'])
chi2_transparency_consent, p_transparency_consent, _, _ = chi2_contingency(contingency_table_transparency_consent)
print(f"\nChi2 Statistic (Transparency Expectation vs. Consent): {chi2_transparency_consent:.2f}")
print(f"P-value (Transparency Expectation vs. Consent): {p_transparency_consent:.3f}")
if p_transparency_consent < 0.05:
    print("Reject the null hypothesis: There is a statistically significant association between transparency expectations and feelings about consent.")
else:
    print("Fail to reject the null hypothesis.")

# 3.b:  Association between Consent and ControlOverAds
contingency_table_consent_control = pd.crosstab(df['Consent'], df['ControlOverAds'])
chi2_consent_control, p_consent_control, _, _ = chi2_contingency(contingency_table_consent_control)
print(f"\nChi2 Statistic (Consent vs. Control Over Ads): {chi2_consent_control:.2f}")
print(f"P-value (Consent vs. Control Over Ads): {p_consent_control:.3f}")
if p_consent_control < 0.05:
    print("Reject the null hypothesis: There is a statistically significant association between feelings about consent and perceived control over ads.")
else:
    print("Fail to reject the null hypothesis.")

# 3.c: Association between Transparency Expectation and PrivacyToolUsage
contingency_table_transparency_privacy = pd.crosstab(df['TransparencyExpectation'], df['PrivacyToolUsage'])
chi2_transparency_privacy, p_transparency_privacy, _, _ = chi2_contingency(contingency_table_transparency_privacy)
print(f"\nChi2 Statistic (Transparency Expectation vs. Privacy Tool Usage): {chi2_transparency_privacy:.2f}")
print(f"P-value (Transparency Expectation vs. Privacy Tool Usage): {p_transparency_privacy:.3f}")
if p_transparency_privacy < 0.05:
    print("Reject the null hypothesis: There is a statistically significant association between transparency expectations and the use of privacy tools.")
else:
    print("Fail to reject the null hypothesis.")

# --- Summary of Findings (will be filled in as you run the code) ---
print("\n--- Summary of Findings ---")
print("Hypothesis 1:")
print("  - Awareness vs. Concern: See above.")
print("  - Bias Noticed vs. Bias Detection: See above.")
print("  - Transparency Expectation (Aware vs. Unaware): See above.")
print("\nHypothesis 2:")
print("  - Bias Detection vs. Manipulation Agreement: See above.")
print("  - Bias Detection vs. Felt Manipulated: See above.")
print("\nHypothesis 3:")
print("  - Transparency Expectation vs. Consent: See above.")
print("  - Consent vs. Control Over Ads: See above.")
print("  - Transparency Expectation vs. Privacy Tool Usage: See above.")
```
